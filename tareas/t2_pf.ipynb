```python
# Programación funcional - Hola Mundo

# Opción 1: Usando una función pura
def saludar(mensaje):
    """Función pura que retorna un saludo"""
    return mensaje

# Usar la función
print(saludar("Hola mundo"))

# Opción 2: Usando funciones de orden superior
def crear_saludo(prefijo):
    """Función de orden superior que retorna otra función"""
    return lambda mensaje: f"{prefijo} {mensaje}"

saludo_español = crear_saludo("Hola")
print(saludo_español("mundo"))

# Opción 3: Usando composición de funciones
def mayusculas(texto):
    return texto.upper()

def agregar_exclamacion(texto):
    return f"{texto}!"

# Composición
mensaje = "hola mundo"
resultado = agregar_exclamacion(mayusculas(mensaje))
print(resultado)
```
